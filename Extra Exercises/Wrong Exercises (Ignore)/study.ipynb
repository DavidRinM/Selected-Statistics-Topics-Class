# Librer√≠as

```python
import pandas as pd
import numpy as np
import datetime as dt

from scikitplot.metrics import plot_roc_curve

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV, train_test_split, cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix, roc_auc_score

from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.neural_network import MLPClassifier
from varclushi import VarClusHi

import pickle

pd.set_option('display.max_columns',500)
```

# Lectura de Datos

```python
df = pd.read_csv("train.csv", parse_dates=["dob"],dtype={
                 "is_fraud": "int8", "merch_lat": "float16", "merch_long":"float16"})
df.drop(columns=["id_txn", "trans_date_trans_time", "cc_num", "merchant", "first", "last", "job", "street", "zip", "city", "lat", "long", "trans_num", "unix_time"],
       axis=1, inplace=True)
df = df.sample(n=100000)
df.info()
```

```python
now = pd.Timestamp("now")
df['dob'] = df['dob'].where(df['dob'] < now, df['dob'] -  np.timedelta64(100, 'Y'))
df['age'] = (now - df['dob']).astype('<m8[Y]')
df.drop(columns=["dob"], inplace=True)
df.head()
```

```python
df.describe(percentiles=(0.01,0.99)).T
```

# Types of Variables

```python
varc = ["amt", "city_pop", "merch_lat", "merch_long", "age"]
vard = [v for v in df if ((v not in varc) and (v != "is_fraud"))]
vard
```

# MLP Classifier

## Preprocessing

```python
aux = df[vard].copy()
oh = OneHotEncoder()
oh.fit(aux)
X = pd.DataFrame(oh.transform(aux).toarray(), columns=oh.get_feature_names())
```

```python
sc = MinMaxScaler()
aux = df[varc].copy()
```

```python
sc.fit(aux)
X[varc] = sc.transform(aux)
```

```python
X.head()
```

## Multicolinealidad

```python
vc = VarClusHi(df=X, maxclus=10)
```

```python
%%time
vc.varclus()
```

```python
r2 = vc.rsquare
r2 = r2.sort_values(by=['Cluster','RS_Ratio'],ascending=[1,1]).reset_index(drop=True)
r2['importancia'] = r2.groupby('Cluster').cumcount()+1
r2
```

```python
best = r2.loc[r2.importancia==1]['Variable'].tolist()
best
```

## Entrenamiento

```python
X = X[best]
y = df["is_fraud"].copy()
```

```python
pi = make_pipeline(PCA(), MinMaxScaler())
```

```python
Xp = pd.DataFrame(pi.fit_transform(X))
```

```python
pi.steps[0][-1].explained_variance_ratio_.cumsum()
```

```python
modelo = MLPClassifier()
```

```python
param = dict(hidden_layer_sizes = [(a,b,) for a in range(1,10)  for b in range(1,10)],
activation = ['identity', 'logistic', 'tanh', 'relu'],
solver = ['lbfgs', 'sgd', 'adam'],
alpha = np.arange(0.0001,0.001,0.0001),
learning_rate = ['constant', 'invscaling', 'adaptive'],
learning_rate_init = np.arange(0.001,0.01,0.001),
max_iter = [len(X)*2])
```

```python
grid = RandomizedSearchCV(estimator=modelo,
                          cv=10,
                          n_iter=300,
                          n_jobs=-1,
                          verbose=True,
                          scoring='roc_auc',
                          param_distributions=param)
```

```python
Xt,Xv,yt,yv = train_test_split(Xp,y,train_size=0.7)
```

```python
grid.fit(Xt,yt)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
modelo = grid.best_estimator_
```

```python
modelo.fit(Xt,yt)
```

```python
plot_roc_curve(y_true=yt,y_probas=modelo.predict_proba(Xt),curves='macro')
plot_roc_curve(y_true=yv,y_probas=modelo.predict_proba(Xv),curves='macro')
```

# Logistic Regression

```python
df_train = pd.read_csv("train.csv", parse_dates=["dob"],dtype={
                 "is_fraud": "int8", "merch_lat": "float16", "merch_long":"float16"})
df_train.drop(columns=["trans_date_trans_time", "cc_num", "merchant", "first", "last", "job", "street", "zip", "city", "lat", "long", "trans_num", "unix_time"],
       axis=1, inplace=True)
```

```python
now = pd.Timestamp("now")
df_train['dob'] = df_train['dob'].where(df_train['dob'] < now, df_train['dob'] -  np.timedelta64(100, 'Y'))
df_train['age'] = (now - df_train['dob']).astype('<m8[Y]')
df_train.drop(columns=["dob"], inplace=True)
```

```python
df_train.head()
```

## Variables declaration

```python
varc = ["amt", "city_pop", "merch_lat", "merch_long", "age"]
vard = [v for v in df_train if ((v not in varc) and (v != "is_fraud") and (v != "id_txn"))]
tgt = "is_fraud"
vard
```

## Encode Discrete Variables

```python
enc = OneHotEncoder(handle_unknown='ignore',sparse=False)
```

```python
enc.fit(df_train[vard])
```

```python
x_vard = enc.transform(df_train[vard])
```

```python
col_vard = enc.get_feature_names(vard)
```

```python
df_train[col_vard] = x_vard
```

##  Merge Encoded Columns with Continuous Variables

```python
train = df_train[['id_txn']+col_vard.tolist()+varc+['is_fraud']]
train.head()
```

## Training Regression

```python
to_train = train.columns[1:-1]
to_train
```

```python
X = train[to_train]
```

```python
y = train['is_fraud']
```

```python
clf = LogisticRegression(random_state=48,class_weight='balanced').fit(X, y)
```

```python
clf.predict_proba(X)
```

## Validating Model

```python
cross_val_score(cv=4,estimator=clf,X=X,y=y,scoring='roc_auc')
```
